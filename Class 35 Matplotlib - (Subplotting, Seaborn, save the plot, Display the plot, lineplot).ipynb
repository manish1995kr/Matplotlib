```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Subplotting

```python
x = np.arange(1,30)
x
```

```python
plt.figure(figsize =(5,7))

plt.subplot(1,4,1)
plt.plot(x,x+1);

plt.subplot(1,4,2)
plt.plot(x,x+2)

plt.subplot(1,4,3)
plt.plot(x,x+3)

plt.subplot(1,4,4)
plt.plot(x,x+4)
```

```python
plt.figure(figsize =(5,7))

plt.subplot(2,2,1)
plt.plot(x,x+1);

plt.subplot(2,2,2)
plt.plot(x,x+2)

plt.subplot(2,2,3)
plt.plot(x,x+3)

plt.subplot(2,2,4)
plt.plot(x,x+4)
```

```python
plt.plot([1,2,3,4,5],[1,4,6,7,3])
plt.ylabel("number")
plt.xlabel("x varaible")
plt.title("list of number")
plt.xticks(np.arange(1,10,1))
plt.yticks(np.arange(1,30,2))
plt.show()
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
plt.plot(df["total_bill"],df["tip"],"o")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("plotting total bill and tip")
plt.xticks(np.arange(10,51,5))
plt.yticks(np.arange(1,12,2));
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")

plt.xlabel('x label')
plt.ylabel('y label')

plt.legend()

plt.show()
```

```python
x = np.arange(1,10,1)

plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")

plt.xlabel('x label')
plt.ylabel('y label')

plt.legend()

plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(1,11,1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show();
```

# line plot

```python
df1 = sns.load_dataset("flights")
df1
```

```python
df1["year"].unique()
```

```python
plt.plot(df1["year"],df1["passengers"],marker = "o")
plt.xlabel("year")
plt.ylabel("passengers")
plt.title("plotting year vs passengers")
plt.xticks(df1["year"].unique());
```

```python
x = df1.groupby(["year"])["passengers"].sum()
x
```

```python
df2 = pd.DataFrame(x)
df2.reset_index(inplace = True)
```

```python
df2
```

```python
plt.plot(df2["year"],df2["passengers"])
```
